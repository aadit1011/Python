```python
import pandas as pd 
import numpy as np 

df=pd.read_csv('matches.csv')
df
```

```python
# Find the city where the last match of every IPL season was played. 

city=df.drop_duplicates(['season'],keep='last')
city[['season','city']].sort_values('season')
```

```python
# Get the team that won the toss in the last match of every season. 

won_toss=df.drop_duplicates('season',keep='last')
won_toss[['season','toss_winner']].sort_values('season')
```

```python
# Identify the umpire1 who officiated the final match of each season.

up1=df.drop_duplicates('season',keep='last')
up1[['season','umpire1']].sort_values('season')
```

```python
# Retrieve the player of the match for the final game in each IPL season

player=df.drop_duplicates('season',keep='last')
player[['season','player_of_match']].sort_values('season')
```

```python
# Find the first match played in each IPL season

first_match=df.drop_duplicates('season')
first_match[['season','team1','team2','winner']].sort_values('season')
```

```python
# Find the city where the first match of every season was played.

first_city=df.drop_duplicates('season')
first_city[['season','city']].sort_values('season')
```

```python
# Retrieve the toss decision (bat/field) in the final match of each IPL season.

toss_decision=df.drop_duplicates('season')
toss_decision[['season','toss_decision']].sort_values('season')
```

```python
# Find the first match where the Duckworth-Lewis method (dl_applied = 1) was applied for every season.


dl_method=df[df['dl_applied']==1]
dl_method=dl_method.drop_duplicates('season',keep='first')
dl_method[['season','team1','team2']].sort_values('season')
```

```python
# Get the last match where the Duckworth-Lewis method was applied in each season.

dl_method=df[df['dl_applied']==1]
dl_method=dl_method.drop_duplicates('season',keep='last')
dl_method[['season','team1','team2']].sort_values('season')
```

```python
# Find the team with the maximum win_by_runs in each season.

maximum=df.sort_values(['win_by_runs'],ascending=False)
maximum.drop_duplicates('season',keep='first')[['season','team1','team2','winner','win_by_runs']].sort_values('season')
```

```python
# Retrieve the team with the maximum win_by_wickets in each IPL season.


maximum=df.sort_values(['win_by_wickets'],ascending=False)
maximum.drop_duplicates('season',keep='first')[['season','team1','team2','winner','win_by_wickets']].sort_values('season')

```

```python
# Get the player who received the most player_of_match awards in the last match of each season.

player=df.drop_duplicates('season',keep='last')
player=player[['season','player_of_match']].sort_values('season')
player.value_counts('player_of_match')
```
