
Creating Function for every batsman

```python
# Function to get maximum runs scored by a batsman against different teams

batsman=input('Enter the name of the batsman')

def maximum_runs_scored_by_batsman(batsman):
    # Filter the DataFrame for the specific batsman
    data = df[df['batsman'] == batsman]
    
    # Group by 'bowling_team' and sum the 'batsman_runs'
    data_grouped = data.groupby('bowling_team').agg({'batsman_runs': 'sum'})
    
    # Sort the resulting Series in descending order of runs
    return data_grouped['batsman_runs'].sort_values(ascending=False)


# Example function call
print(maximum_runs_scored_by_batsman(batsman))
```
