```python
# Retrieve the team1 and team2 for the first 10 matches played in Hyderabad 

hyderabad=df[df['city']=='Hyderabad']
hyderabad[['team1','team2']].iloc[:10]

```

```python
# Get the winner, win_by_runs, and win_by_wickets for the first 20 matches where Royal Challengers Bangalore played. 

rcb=df[(df['team1']=='Royal Challengers Bangalore') | (df['team2']=='Royal Challengers Bangalore')]
rcb[['winner','win_by_runs','win_by_wickets']].iloc[:20]
```

```python
# Fetch umpire1, umpire2, and umpire3 for all matches with a dl_applied value of 1. 

dl_applied=df[df['dl_applied']==1]
dl_applied[['umpire1','umpire2','umpire3']]
```

```python
# Retrieve the entire data for matches where Mumbai Indians won and use iloc for specific rows.

mi=df[df['winner']=='Mumbai Indians']
mi.iloc[2:30]
```

```python
# Get match data for rows corresponding to matches where the win_by_wickets is greater than 5 using iloc.  

greater=df[df['win_by_wickets']>5]
greater.iloc[2:30]
```

```python
# Fetch the date, city, and winner for all matches where the toss_decision was "bat" using iloc. 

toss_decision=df[df['toss_decision']=='bat']
toss_decision[['date','city','winner']].iloc[:31]
```

```python
# Retrieve the team1, team2, and player_of_match for matches that happened in 2017 and had more than 100 runs victory using iloc.

match2017=df[(df['date']>='2017-01-01') & (df['date']<'2018-01-01' )]
more_than_100=match2017[match2017['win_by_runs']>100]
more_than_100[['team1','team2','player_of_match']]
```

```python
# Get the venue, city, date, and winner for the first 5 matches that Sunrisers Hyderabad won using iloc.

srh=df[df['winner']=='Sunrisers Hyderabad']
srh[['date','venue','city','winner']].iloc[:5]
```

```python
# Fetch the toss_winner, toss_decision, and winner for all matches where the winner is the same as toss_winner using iloc

same=df[df['toss_winner']==df['winner']]
same[['toss_winner','toss_decision','winner']]
```

```python
# Retrieve match data for rows corresponding to the top 10 matches with the highest win_by_runs values using iloc. 

sorted_values=df.sort_values(by='win_by_runs',ascending=False)

sorted_values.iloc[:10]
```
