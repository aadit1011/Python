```python
import pandas as pd 
# How can you filter the dataset to show only movies released in the year 1999?
df=pd.read_csv('movie_titles_metadata.tsv',sep='\t',names=['id','Movie_name','Date_of_Release','Ratings','Voting','Genre'],index_col=['id'])

year_1999=df[df['Date_of_Release']=='1999']
year_1999
```

```python
# Find the number of movies with a rating above 7.5.

movies_above_7p5=df[df['Ratings']>7.5]
movies_above_7p5
```

```python
# Display all movies released before 1980.
released_before_1980=df[df['Date_of_Release']<'1980']
released_before_1980
```

```python
# Find the movie(s) with the highest rating.

highest=df['Ratings'].max()
movie_name=df[df['Ratings']==highest] 
movie_name
```
